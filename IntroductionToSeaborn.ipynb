#### Adding a third variable with *hue*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')

hue_colors = {'Yes': 'black', 'No': 'red'}

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker',
                hue_order=['Yes','No'],
                palette=hue_colors)

plt.show()
```

```python
sns.countplot(x='smoker',data=tips,hue='sex')

plt.show()
```

### Relational plots and subplots

#### Introducing relplot()

- Create 'relational plots': scatter plots or lineplots?
- Lets you create subplots in a single figure.


```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter')

plt.show()
```

- Subpots in columns:

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='smoker')

plt.show()
```

- Subplots in rows:

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',row='smoker')

plt.show()
```

- Subplots in rows and columns:

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='smoker',row='time')

plt.show()
```

- Wrapping columns:

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='day',col_wrap=2)

plt.show()
```

- Ordering columns:

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='day',col_wrap=2,col_order=['Thur','Fri','Sat','Sun'])

plt.show()
```

- Subgroups with point size

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',size='size',hue='size')
plt.show()
```

- Subgroups with point style

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',style='smoker',hue='smoker')

plt.show()
```

- Changing point transparency

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',alpha=0.4)

plt.show()
```

### Categorical plots

- Examples: count plots, bar plots
- Involve a categorical variable
- Comparison between groups

#### catplot()

- Used to create categorical plots
- Same advantages of relplot()
- Easily create subplots with col and row

~~~
sns.catplot(x='how_masculine',data=masculinity_data,kind='count')
~~~

#### Changing the order

~~~
category_order = ['No answer','Not at all','Not very',
			'Somewhat','Very']

sns.catplot(x='how_masculine',data=masculinity_data
		kind='count',order=category_order)
~~~

#### Bar plots

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
plt.show()
```

- Confidence intervals:
	- Lines show 95% confidence intervals for the mean.
	- Shows uncertainty about our estimate.
	- Assumes our data is a random sample.

#### Turning off CI

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',ci=None)
plt.show()
```

#### Changing the orientation

```python
sns.catplot(x='total_bill',y='day',data=tips,kind='bar')
plt.show()
```

When the y-variable is True/False, bar plots will show the percentage of responses reporting True.

### Box plots

- Show the distribution of quantitative data.
- See median, spread, skewness and outliers.
- Facilitates comparisons between groups.

```python
g = sns.catplot(x='time',y='total_bill',data=tips,kind='box',order=['Dinner','Lunch'])
plt.show()
```

#### Omitting outliers

```python
g = sns.catplot(x='time',y='total_bill',data=tips,kind='box',sym='')
plt.show()
```

#### Changing the whiskers

- By default, 1.5 * IQR
- Extend to 2 * IQR:

```python
g = sns.catplot(x='time',y='total_bill',data=tips,kind='box',whis=2.0)
plt.show()
```

- Show the 5th and 95th percentiles:

```python
g = sns.catplot(x='time',y='total_bill',data=tips,kind='box',whis=[5, 95])
plt.show()
```

### Point plots

- Show mean of quantitative variable.
- Vertical lines show 95% CI.

~~~

#### Displaying the median

```python
from numpy import median

sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',estimator=median)

plt.show()
```

#### Customizing the CIs

```python
sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',capsize=0.2)

plt.show()
```

#### Turning off CI

```python
sns.catplot(x='smoker',y='total_bill',data=tips,kind='point',ci=None)

plt.show()
```

#### Point plots vs. line plots

- Both show:
	- Mean of quantitative variable.
	- 95% CI for the mean.
- Differences:
	- Line plot has **quantitative** variable on the x-axis.
	- Point plot has **categorical** variable on the x-axis.

#### Point plots vs. bar plots

- Both show:
	- Mean of quantitative variable.
	- 95% CI for the mean.

### Adding titles and labels

- Seaborn plots create two different types of objects: FacetGrid and AxesSubplot

~~~
g = sns.scatterplot(x='height',y='weight',data=df)

type(g)
~~~

- FacetGrid consists of one or more AxesSubplots (which is how it supports subplots).

| Object Type  | Plot Types                       | Characteristics          |
|--------------|----------------------------------|--------------------------|
| FacetGrid    | catplot(), relplot()             | Can create subplots      |
| AxesSubplots | scatterplot(), countplot(), etc. | Only create single plots |

#### Adding a title to a FacetGrid

~~~
g = sns.catplot(x='Region',y='Birthrate',data=gdp_data,kind='box')

g.fig.suptitle('New Title', y=1.03)
~~~

#### Adding a title to AxesSubplot

~~~
g = sns.catplot(x='Region',y='Birthrate',data=gdp_data,kind='box')

g.set_title('New Title', y=1.03)
~~~

#### Titles for subplots

~~~
g = sns.catplot(x='Region',y='Birthrate',data=gdp_data,kind='box')

g.fig.suptitle('New Title', y=1.03)

g.set_title('This is {col_name}')
~~~

#### Adding axis labels

~~~
g = sns.catplot(x='Region',y='Birthrate',data=gdp_data,kind='box')

g.set(xlabel='New X Label',ylabel='New Y Label')

plt.show()
~~~
